```python
import FloatSatEnv
import numpy as np

import matplotlib.pyplot as plt


angles = np.linspace(-180, 180, 1000000)
rewards = []
for angle in angles:
    rewards.append(FloatSatEnv.calc_angle_reward(angle))   

#Plot result
plt.plot(angles, rewards)
# show zero line
plt.axhline(0, color='black', lw=1)
plt.xlabel('Angle [rad]')
plt.ylabel('Reward')
plt.title('Reward for Angle')
plt.show()

```

```python
angles = np.linspace(-200, 200, 1000)
rewards = []
for angle in angles:
    rewards.append(FloatSatEnv.calc_angular_velocity_reward(angle))   

#Plot result
plt.plot(angles, rewards)
# show zero line
plt.axhline(0, color='black', lw=1)
plt.xlabel('Velocity')
plt.ylabel('Reward')
plt.title('Reward for Angular Velocity')
plt.show()
```

```python
# 2d array with angles and targets pairs
angles_and_targets = np.random.uniform(-np.pi/2, np.pi/2, (10, 2))
# angles_and_targets = [[-np.pi/4, 0]]


# 10 polar plots with line between angle and target
fig = plt.figure()
for i, [angle, target] in enumerate(angles_and_targets):
    # get subplot
    ax = fig.add_subplot(2, 5, i+1, polar=True)

    error = FloatSatEnv.calc_angle_error(np.degrees(angle), np.degrees(target))
    #plot arrow between angle and target
    print(angle, target)
    print(error)
    r = range(0, int(error)) if error > 0 else range(int(error), 0)
    for j in r:
        print(np.radians(angle-j))
        ax.plot(np.radians(angle-j), 1, 'xr')

    ax.plot(angle, 1, marker='o')
    ax.plot(target, 1, marker='o')
plt.show()
```
